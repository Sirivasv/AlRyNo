# Functions

## Get Binary ROC values

```python
def get_sim_after_threshold(input_sim_values, threshold_i):
    result_list_mod = []
    for sim_value_i in input_sim_values:
        if (sim_value_i >= threshold_i):
            result_list_mod.append(1)
        else:
            result_list_mod.append(0)
    return result_list_mod
```

```python
def get_binary_ROC_values(scores_metric_and_experts_LOADED, flip_values=1.0, value_to_get=0, scale=0):
    raw_scorelist_sim_measure_BI = []
    raw_scorelist_sim_measure_majority_BI = []
    raw_scorelist_sim_exp_1_BI = []
    raw_scorelist_sim_exp_2_BI = []
    raw_scorelist_sim_exp_3_BI = []
    raw_scorelist_sim_exp_majority_BI = []

    for keyname_i in scores_metric_and_experts_LOADED.keys():
        for el in scores_metric_and_experts_LOADED[keyname_i]:
            
            sim_score = el[0] * flip_values
            exp_score_1 = el[1]
            exp_score_2 = el[2]
            exp_score_3 = el[3]
            
            raw_scorelist_sim_measure_BI.append(sim_score)
            raw_scorelist_sim_exp_1_BI.append(exp_score_1)
            raw_scorelist_sim_exp_2_BI.append(exp_score_2)
            raw_scorelist_sim_exp_3_BI.append(exp_score_3)

            # majority
            annotators_scores_count = dict(Counter([exp_score_1, exp_score_2, exp_score_3]))
            (max_count, max_value) = max([ (annotators_scores_count[keyname_ann], keyname_ann) for keyname_ann in annotators_scores_count ])
            
            # Majority
            # if max_count == 1:
            #    continue
            
            # Unanimity
            if max_count != 3:
                continue
            
            # SCALE
            valid = True
            if (scale == 0): # 0 - 0.5
                if (max_value == 1):
                    valid = False
                elif (max_value == 0.5):
                    max_value = 1
                elif (max_value == 0):
                    max_value = 0
            elif (scale == 1): # 0.5 - 1.0
                if (max_value == 0):
                    valid = False
                elif (max_value == 1):
                    max_value = 1
                elif (max_value == 0.5):
                    max_value = 0
            elif (scale == 2): # 0 - 1.0
                if (max_value == 0.5):
                    valid = False
                elif (max_value == 1):
                    max_value = 1
                elif (max_value == 0):
                    max_value = 0
            
            if (not valid):
                continue
            
            raw_scorelist_sim_exp_majority_BI.append(max_value)
            raw_scorelist_sim_measure_majority_BI.append(sim_score)
            

    fpr = None
    tpr = None
    precision = None
    recall = None
    thresholds = None
    thresholds_roc = None
    
    print("Found #{0} elements in the selected scale range".format(len(raw_scorelist_sim_exp_majority_BI)))
    if (value_to_get == 0): # Return Majority Vote
        
        fpr, tpr, thresholds_roc = roc_curve(raw_scorelist_sim_exp_majority_BI, raw_scorelist_sim_measure_majority_BI)
        precision, recall, thresholds = precision_recall_curve(raw_scorelist_sim_exp_majority_BI, raw_scorelist_sim_measure_majority_BI)
        
    elif (value_to_get == 1): # Return Expert 1 Vote
        
        fpr, tpr, thresholds_roc = roc_curve(raw_scorelist_sim_exp_1_BI, raw_scorelist_sim_measure_BI)
        precision, recall, thresholds = precision_recall_curve(raw_scorelist_sim_exp_1_BI, raw_scorelist_sim_measure_BI)
        
    elif (value_to_get == 2): # Return Expert 2 Vote
        
        fpr, tpr, thresholds_roc = roc_curve(raw_scorelist_sim_exp_2_BI, raw_scorelist_sim_measure_BI)
        precision, recall, thresholds = precision_recall_curve(raw_scorelist_sim_exp_2_BI, raw_scorelist_sim_measure_BI)
        
    elif (value_to_get == 3): # Return Expert 3 Vote
        
        fpr, tpr, thresholds_roc = roc_curve(raw_scorelist_sim_exp_3_BI, raw_scorelist_sim_measure_BI)
        precision, recall, thresholds = precision_recall_curve(raw_scorelist_sim_exp_3_BI, raw_scorelist_sim_measure_BI)
        
    roc_auc = auc(fpr, tpr)
    
    # print(len(precision))
    # print(len(recall))
    # print(len(thresholds))
    # print(len(fpr))
    # print(len(tpr))
    # print(len(thresholds_roc))
    
    min_diff = 100
    min_diff_thd = None
    min_diff_id = None
    for thd_i in range(len(thresholds_roc)):
        current_diff = fpr[thd_i] + (1.0 - tpr[thd_i])
        if current_diff < min_diff:
            min_diff = current_diff
            min_diff_thd = thresholds_roc[thd_i]
            min_diff_id = thd_i
            
    sim_values_with_threshold_roc = get_sim_after_threshold(raw_scorelist_sim_measure_majority_BI, min_diff_thd)
    tn, fp, fn, tp = confusion_matrix(raw_scorelist_sim_exp_majority_BI, sim_values_with_threshold_roc).ravel()
    
    best_roc_value = 1.0 - min_diff
    best_roc_f1 = (2.0 * tp) / ((2.0 * tp) + fp + fn)
    best_roc_precision = tp / (tp + fp)
    best_roc_recall = tp / (tp + fn)
    best_roc_specificity = tn / (tn + fp)
    
    
    # F1 Score
    numerator = 2 * recall * precision
    denom = recall + precision
    f1_scores = np.divide(numerator, denom, out=np.zeros_like(denom), where=(denom!=0))
    max_f1 = np.max(f1_scores)
    max_fi_id = np.argmax(f1_scores)
    max_f1_thresh = thresholds[max_fi_id]
    max_f1_precision = precision[max_fi_id]
    max_f1_recall = recall[max_fi_id]
    sim_values_with_threshold = get_sim_after_threshold(raw_scorelist_sim_measure_majority_BI, max_f1_thresh)
    tn, fp, fn, tp = confusion_matrix(raw_scorelist_sim_exp_majority_BI, sim_values_with_threshold).ravel()
    max_f1_specificity = tn / (tn + fp)
    
    # print(max_f1_thresh)
    
    # Specifity
    max_specificity = -10
    max_specificity_thresh = None
    # for threshold_i in thresholds:
    #     sim_values_with_threshold = get_sim_after_threshold(raw_scorelist_sim_measure_majority_BI, threshold_i)
    #     tn, fp, fn, tp = confusion_matrix(raw_scorelist_sim_exp_majority_BI, sim_values_with_threshold).ravel()
    #     current_specificity = tn / (tn + fp)
    #     if current_specificity > max_specificity:
    #         max_specificity = current_specificity
    #         max_specificity_thresh = threshold_i
    
    thresholds = np.append(thresholds,-1)
    
    # Precision
    max_precision = np.max(precision)
    max_precision_thresh = thresholds[np.argmax(precision)]
    
    # Recall
    max_recall = np.max(recall)
    max_recall_thresh = thresholds[np.argmax(recall)]
    
    return {
        "fpr":fpr,
        "tpr":tpr,
        "auc":roc_auc,
        "max_precision":max_precision,
        "max_precision_thresh":max_precision_thresh,
        "max_recall":max_recall,
        "max_recall_thresh":max_recall_thresh,
        "max_f1":max_f1,
        "max_f1_thresh":max_f1_thresh,
        "max_f1_precision":max_f1_precision,
        "max_f1_recall":max_f1_recall,
        "max_f1_specificity":max_f1_specificity,
        "max_specificity":max_specificity,
        "max_specificity_thresh":max_specificity_thresh,
        "best_roc_value":best_roc_value,
        "best_roc_f1":best_roc_f1,
        "best_roc_precision":best_roc_precision,
        "best_roc_recall":best_roc_recall,
        "best_roc_specificity":best_roc_specificity
    }
```
